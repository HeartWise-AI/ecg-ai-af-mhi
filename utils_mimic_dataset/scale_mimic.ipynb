```python
import pandas as pd
data_mimic = pd.read_csv("/media/data1/ravram/MIMIC-IV/mimic_index.corrected.csv")
data_mimic['patient_id'] = [int(i.split('/')[8].split('p')[-1]) for i in data_mimic['waveform_path']]
data_mimic
```

```python
# Open the file and read all lines except the first one
with open('/media/data1/anolin/pred_waveform_path.csv', 'r') as file:
    lines = file.readlines()[1:]  # Skip the first line

# Remove any trailing newline characters from each line
lines = [line.strip() for line in lines]

# Print the list to verify
data_mimic = data_mimic[data_mimic['waveform_path'].isin(lines)]
data_mimic
```

```python
import pandas as pd
import numpy as np
from tqdm import tqdm
from concurrent.futures import ThreadPoolExecutor, as_completed

# Function to load and subsample the numpy array, and convert to float16
def load_and_subsample(path):
    arr = np.expand_dims(np.load(path)[1::2, :].astype(np.float16), axis=0)  # Subsample to (2500, 12) by taking every second row and convert to float16
    return arr

# Load the DataFrame (replace with actual loading method if needed)
# data_mimic = pd.read_csv('your_file.csv')

# List to store results from the futures
results = []

# Use ThreadPoolExecutor to parallelize the loading process
with ThreadPoolExecutor() as executor:
    futures = {executor.submit(load_and_subsample, path): path for path in data_mimic['waveform_path']}
    for future in tqdm(as_completed(futures), total=len(futures), desc="Loading and subsampling waveforms"):
        results.append(future.result())

# Concatenate all arrays into a single numpy array
all_waveforms = np.vstack(results)

# Print the shape of the resulting array
print("Shape of the concatenated array:", all_waveforms.shape)
```

```python
# Function to check for non-numerical values
def is_numerical(arr):
    return np.isfinite(arr).all()

# Filter out arrays with non-numerical values using tqdm for progress
all_waveforms_filtered = np.array([arr for arr in tqdm(all_waveforms, desc="Filtering non-numerical arrays") if is_numerical(arr)])
```

```python
from sklearn.preprocessing import StandardScaler
import joblib

# Flatten the array for mean and std computation
flattened_waveforms = all_waveforms_filtered.reshape(-1, 1)

# Compute the mean and standard deviation
mean = np.nanmean(flattened_waveforms)
std = np.nanstd(flattened_waveforms)

scaler = StandardScaler()
scaler.mean_ = np.array([mean])
scaler.scale_ = np.array([std])
scaler.var_ = np.array([std**2])
scaler.n_features_in_ = 1

# Save the scaler object using joblib
joblib.dump(scaler, 'scaler.pkl')
```
